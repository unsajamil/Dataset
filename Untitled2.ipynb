```python
missing_data = []

for name, df in datasets.items():
    for col in df.columns:
        missing_count = df[col].isnull().sum()
        if missing_count > 0:
            missing_data.append({
                "Dataset": name,
                "Column": col,
                "Missing Count": missing_count,
                "Missing Percentage": round((missing_count / len(df)) * 100, 2)
            })

missing_df = pd.DataFrame(missing_data)
print("Missing Values Summary\n")
print(missing_df.to_string(index=False))
```

#5.Data Aggregation & Joining

##5.1 Storage Framework Selection

**Selected Framework: Apache Spark with Parquet Storage**

Luxora Hotels generates large volumes of data daily across 6 different systems including room bookings, guest profiles, staff scheduling, maintenance logs, guest feedback and in-room service usage. After evaluating several frameworks, Apache Spark with Parquet storage has been selected for the following reasons:

1. Scalability: Apache Spark is designed for large scale distributed data processing and can handle millions of records across multiple hotel branches in the MENA region.
2. Speed: Spark processes data in-memory which makes it significantly faster than traditional MapReduce frameworks for both batch and real-time analytics.
3. Parquet Format: Parquet is a columnar storage format that compresses data efficiently and allows fast query execution on specific columns without reading the entire dataset.
4. SQL Support: Spark SQL allows analysts to run familiar SQL queries on large datasets making it easier for Luxora's analytics team to extract insights.
5. Integration: Spark integrates seamlessly with Hadoop HDFS, cloud storage like AWS S3, and machine learning libraries making it future-proof for Luxora's growing needs.
6. Fault Tolerance: Spark's RDD architecture ensures data is not lost in case of failures which is critical for hotel operations running 24/7.

#6. Exploratory Data Analysis (EDA)

##6.1 Define Hypothesis

**Hypothesis**

Guests with higher loyalty tiers (Gold and Platinum) tend to spend more on in-room services and give higher satisfaction ratings compared to guests with lower loyalty tiers (Bronze and Silver).

**Why this is interesting**

Luxora Hotels runs a loyalty program called LuxRewards with 4 tiers — Bronze, Silver, Gold and Platinum. Understanding whether higher tier guests spend more and are more satisfied can help Luxora:



*   Target high value guests with personalized offers
*   Improve services for lower tier guests to retain them
*   Allocate resources based on guest loyalty segments
*   Design better reward programs to increase guest spending

**Datasets Used**

*   Customer Profiles: contains loyalty_tier for each guest
*   In-Room Service Usage: contains cost of services used by each guest
*   Guest Feedback: contains rating given by each guest

**Expected Outcome**


*   Gold and Platinum guests will show higher average service costs
*   Gold and Platinum guests will show higher average satisfaction ratings
*   Bronze and Silver guests will show lower spending and lower ratings

##6.2 Statistical Analysis

```python
customer_service = pd.merge(
    customer_profiles,
    inroom_service_usage,
    on="customer_id",
    how="inner"
)

customer_feedback = pd.merge(
    customer_profiles,
    guest_feedback,
    on="customer_id",
    how="inner"
)

avg_spending = customer_service.groupby("loyalty_tier", observed=True)["cost"].mean().reset_index()
avg_spending.columns = ["Loyalty Tier", "Average Service Cost"]
avg_spending["Average Service Cost"] = avg_spending["Average Service Cost"].round(2)

avg_rating = customer_feedback.groupby("loyalty_tier", observed=True)["rating"].mean().reset_index()
avg_rating.columns = ["Loyalty Tier", "Average Rating"]
avg_rating["Average Rating"] = avg_rating["Average Rating"].round(2)

service_count = customer_service.groupby("loyalty_tier", observed=True)["service_id"].count().reset_index()
service_count.columns = ["Loyalty Tier", "Total Services Used"]

final_analysis = avg_spending.merge(avg_rating, on="Loyalty Tier").merge(service_count, on="Loyalty Tier")

print("Loyalty Tier Analysis Summary\n")
display(final_analysis.style.format({
    "Average Service Cost": "{:.2f}",
    "Average Rating": "{:.2f}"
}).set_table_styles([
    {"selector": "th", "props": [("border", "1px solid black"), ("padding", "8px"), ("background-color", "#f2f2f2"), ("text-align", "center")]},
    {"selector": "td", "props": [("border", "1px solid black"), ("padding", "8px"), ("text-align", "center")]},
    {"selector": "table", "props": [("border-collapse", "collapse"), ("width", "100%")]}
]))
```

##6.6 Findings

**EDA Findings Summary**

**1. Loyalty Tier vs Spending and Rating**

Our analysis revealed an unexpected pattern in guest behavior across loyalty tiers. Bronze tier guests recorded the highest average service cost of 431.87 and were the most active users of in-room services with a total of 285 service requests. This suggests that Bronze tier guests are highly engaged with hotel amenities despite being at the lowest loyalty level. Platinum tier guests had the second highest average spending of 430.80 however they gave the lowest satisfaction ratings of 5.26 which is concerning for Luxora as these are their most loyal customers. Silver tier guests gave the highest average rating of 6.07 despite having the lowest average spending of 386.63 which indicates that spending does not always correlate with satisfaction.

**2. Most Common Maintenance Issue**

The maintenance logs analysis revealed that certain issues are more frequent than others across all rooms in Luxora Hotels. Addressing these recurring issues proactively can significantly improve guest satisfaction and reduce operational costs. A preventive maintenance schedule should be implemented for the most common issues to minimize disruptions to guests during their stay.

**3. Most Popular Service Type**

Room Service emerged as the most popular in-room service across all loyalty tiers. Bronze guests used Room Service 120 times while Platinum guests used it 45 times. This consistent preference for Room Service across all tiers indicates that Luxora should invest in improving and expanding its Room Service offerings. Faster delivery times and a wider menu selection could further enhance guest satisfaction across all loyalty segments.

**4. Booking Trends by Room Type**

Standard rooms dominated bookings with 294 total reservations and an average price per night of 825.54 making them the primary revenue driver for Luxora Hotels. Deluxe rooms came in second with 74 bookings at an average price of 1023.51 per night. Presidential rooms had the second highest average price of 962.21 but relatively low bookings at 67 suggesting they cater to a niche high end market. Suite rooms had the fewest bookings at 65 with an average price of 898.45 per night indicating potential for targeted marketing campaigns to increase suite occupancy rates.

**5. Overall Business Insight**

The EDA results highlight several key areas where Luxora Hotels can improve its operations and guest experience. The gap between Platinum guest spending and their satisfaction ratings suggests that the hotel needs to reassess the quality of services provided to its most loyal guests. The dominance of Room Service across all loyalty tiers presents an opportunity for Luxora to differentiate itself through premium room service offerings. The high volume of Standard room bookings indicates strong demand in the mid-range segment which Luxora should continue to nurture while also developing strategies to increase bookings for higher priced room types.
