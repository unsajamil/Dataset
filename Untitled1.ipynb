#4.Completeness — Missing Value Detection and Treatment

##4.1 Missing Values Count and Percentage Per Column

```python
print(df.isnull().sum())
```

```python
print((df.isnull().sum() / len(df)) * 100)
```

##4.2 Missing Value Heatmap and Percentage Bar Chart across all Columns

```python
plt.figure(figsize=(9, 5))
sns.heatmap(df.isnull(), cbar=True, cmap='viridis', yticklabels=False)
plt.title('Missing Value Heatmap Across all Columns')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
missing_percent = (df.isnull().sum() / len(df)) * 100
missing_percent = missing_percent[missing_percent > 0]
plt.figure(figsize=(10, 5))
missing_percent.plot(kind='bar', color='tomato')
plt.title('Missing Value Percentage per Column')
plt.ylabel('Missing %')
plt.tight_layout()
plt.show()
```

##4.3 Filling Missing Numerical Values with Median

```python
import warnings
warnings.filterwarnings('ignore')

numerical_cols = df.select_dtypes(include=np.number).columns
for col in numerical_cols:
    if df[col].isnull().sum() > 0:
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)
        print(f"{col} filled with median: {median_val}")
```

##4.4 Filling Missing Categorical Values with Mode

```python
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        mode_val = df[col].mode()[0]
        df[col].fillna(mode_val, inplace=True)
        print(f"{col} filled with mode: {mode_val}")
```

##4.5 Dropping Columns with More Than 50% Missing Values

```python
threshold = 50
cols_to_drop = [col for col in df.columns if (df[col].isnull().sum() / len(df)) * 100 > threshold]
if cols_to_drop:
    df.drop(columns=cols_to_drop, inplace=True)
    print(f"Dropped columns: {cols_to_drop}")
else:
    print("No columns exceeded 50% missing threshold - none dropped")

print("\nMissing Values After Cleaning\n")
print(df.isnull().sum())
```
