#2.Data Cleaning

##2.1 Fill and Drop Missing Values

```python
df = df.dropna(subset=['diagnosis', 'treatment_plan'])

df['gender'].fillna('Unknown', inplace=True)
df['blood_type'].fillna('Unknown', inplace=True)
df['insurance_provider'].fillna('Unknown', inplace=True)
df['doctor_id'].fillna(0, inplace=True)
df['nurse_id'].fillna(0, inplace=True)

print("Missing values after cleaning:")
print(df.isnull().sum())
```

##2.2 Convert Date Columns

```python
df['admission_date'] = pd.to_datetime(df['admission_date'])
df['discharge_date'] = pd.to_datetime(df['discharge_date'])

df['length_of_stay'] = (df['discharge_date'] - df['admission_date']).dt.days

df = df[df['length_of_stay'] >= 0]

print(df[['admission_date', 'discharge_date', 'length_of_stay']].head())
print("\nLength of Stay Stats:")
print(df['length_of_stay'].describe())
print("\nDataset shape after fix:", df.shape)
```

##2.3 Extract Age from DOB

```python
df['dob'] = pd.to_datetime(df['dob'])
df['age'] = (pd.Timestamp.now() - df['dob']).dt.days // 365

print(df[['first_name', 'dob', 'age']].head())
```

#4.SQL Analysis

##4.1 Setup SQL

```python
import sqlite3

conn = sqlite3.connect(':memory:')
df.to_sql('patients', conn, index=False, if_exists='replace')
print("SQL Database created successfully!")
```

##4.2 Top Diagnoses by Count

```python
query1 = """
SELECT diagnosis, COUNT(*) as total_patients
FROM patients
GROUP BY diagnosis
ORDER BY total_patients DESC
"""
result1 = pd.read_sql_query(query1, conn)
print(result1)
```

##4.3 Average Length of Stay by Diagnosis

```python
query2 = """
SELECT diagnosis,
       ROUND(AVG(length_of_stay), 2) as avg_stay_days,
       COUNT(*) as total_patients
FROM patients
GROUP BY diagnosis
ORDER BY avg_stay_days DESC
"""
result2 = pd.read_sql_query(query2, conn)
print(result2)
```

##4.4 Most Common Treatment per Diagnosis

```python
query3 = """
SELECT diagnosis, treatment_plan, COUNT(*) as total
FROM patients
GROUP BY diagnosis, treatment_plan
ORDER BY diagnosis, total DESC
"""
result3 = pd.read_sql_query(query3, conn)
print(result3)
```

##4.5 Elderly Patients (81+) Analysis

```python
query4 = """
SELECT diagnosis, COUNT(*) as elderly_patients
FROM patients
WHERE age > 80
GROUP BY diagnosis
ORDER BY elderly_patients DESC
"""
result4 = pd.read_sql_query(query4, conn)
print(result4)
```

##4.6 Gender & Diagnosis Breakdown

```python
query5 = """
SELECT gender, diagnosis, COUNT(*) as total
FROM patients
WHERE gender IN ('Male', 'Female')
GROUP BY gender, diagnosis
ORDER BY gender, total DESC
"""
result5 = pd.read_sql_query(query5, conn)
print(result5)
```
